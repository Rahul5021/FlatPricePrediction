```python
# Reset the index after dropping duplicates
df.reset_index(drop=True, inplace=True)
```

```python
import re

sale_price_cr = []
for i in range(df.shape[0]):
    # Access the value using iloc to avoid KeyError
    amount = df.iloc[i]["Amount(in rupees)"]
    
    # Check if there's any numeric value in the string (handles decimals, commas, etc.)
    if bool(re.search(r'\d', amount)):  # If there is any digit in the string
        # Split the amount into value and unit
        split_amount = amount.split()
        
        # Check if the unit is 'lac'
        if len(split_amount) == 2 and split_amount[1].lower() == 'lac':
            # Convert the amount to float and divide by 100 for 'lac'
            sale_price_cr.append(float(split_amount[0]) / 100)
        else:
            # Otherwise, just append the value as it is (convert to float)
            sale_price_cr.append(float(split_amount[0]))
    else:
        # If no numeric value is found (e.g., 'Call for Price'), set to NaN
        sale_price_cr.append(np.nan)
```

```python
sum([1 for value in sale_price_cr if pd.isna(value)])
```

```python
df["SalePrice(in Crores)"] = sale_price_cr
```

```python
calculated_area = []

for i in range(df.shape[0]):
    # Extract values
    sale_price_in_rupees = df.iloc[i]["SalePrice(in Crores)"] * 10**7  # Convert crores to rupees
    price_per_sqft = df.iloc[i]["Price (in rupees)"]
    
    # Check if either value is NaN
    if np.isnan(sale_price_in_rupees) or np.isnan(price_per_sqft) or price_per_sqft == 0:
        # If NaN or price is zero, append NaN or handle the situation as needed
        calculated_area.append(np.nan)  # Or append 0, or skip the row if preferred
    else:
        # Calculate the area, round it, and convert to integer
        calculated_area.append(int(round(sale_price_in_rupees / price_per_sqft)))
```

```python
import re
import numpy as np

# Function to extract the number of BHK
def extract_bhk(title):
    # Search for the pattern of number followed by 'bhk'
    match = re.search(r'(\d+)\s*bhk', title.lower())  # Convert title to lowercase to avoid case issues
    if match:
        return int(match.group(1))  # Return the number of BHK as integer
    else:
        return np.nan  # Return NaN if no match is found

# Apply the function to the 'Title' column and create a new 'BHK' column
df['BHK'] = df['Title'].apply(extract_bhk)

# Check the results
print(df[['Title', 'BHK']].head())
```

```python
carpet_area = []
for i in range(df.shape[0]):
    # Access the value using iloc to avoid KeyError
    c_area = df.iloc[i]["Carpet Area"]
    
    # Convert c_area to string to handle both numeric and non-numeric values
    c_area_str = str(c_area)
    
    # Check if there's any numeric value in the string (handles decimals, commas, etc.)
    if bool(re.search(r'\d', c_area_str)):  # If there is any digit in the string
        # Split the amount into value and unit
        split_c_area = c_area_str.split()
        
        # Check if the unit is 'sqft'
        if len(split_c_area) == 2 and split_c_area[1].lower() == 'sqft':
            # Convert the amount to float and append only the numeric value
            carpet_area.append(float(split_c_area[0]))
        else:
            # If the unit is not recognized, append NaN or a default value
            carpet_area.append(np.nan)
    else:
        # If no numeric value is found (e.g., 'Call for Price'), set to NaN
        carpet_area.append(np.nan)
```
