```python
import re
import numpy as np

flat_floor = []
total_floors = []
error = []
basement = []

for i in range(df.shape[0]):
    floor = df.iloc[i]["Floor"]
    floor_str = str(floor).lower()

    try:
        # Split the floor entry into parts
        split_floor = floor_str.split()

        # Handle "basement" cases
        if 'basement' in split_floor:
            basement.append(True)
            if 'upper' in split_floor:
                flat_floor.append(-1)  # Upper basement
            elif 'lower' in split_floor:
                flat_floor.append(-2)  # Lower basement
            else:
                flat_floor.append(-1)  # Default basement level
            # Extract total floors if present
            total_floors.append(
                int(split_floor[-1]) if len(split_floor) > 1 and re.search(r'\d', split_floor[-1]) else np.nan
            )

        # Handle "ground floor" cases
        elif 'ground' in split_floor:
            basement.append(False)
            flat_floor.append(0)  # Ground level
            # Extract total floors if present
            total_floors.append(
                int(split_floor[-1]) if len(split_floor) > 1 and re.search(r'\d', split_floor[-1]) else np.nan
            )

        # Handle numeric floors
        elif re.search(r'\d', floor_str):
            basement.append(False)
            flat_floor.append(int(re.search(r'\d+', split_floor[0]).group()))  # Extract numeric floor
            total_floors.append(
                int(split_floor[-1]) if len(split_floor) > 1 and re.search(r'\d', split_floor[-1]) else np.nan
            )

        # Log unexpected formats
        else:
            flat_floor.append(np.nan)
            total_floors.append(np.nan)
            basement.append(False)
            error.append(split_floor)

    except Exception as e:
        # Handle unforeseen parsing errors
        flat_floor.append(np.nan)
        total_floors.append(np.nan)
        basement.append(False)
        error.append(f"Error parsing row {i}: {floor_str} - {e}")
```

```python
# Function to check if a specific feature exists in 'overlooking'
def check_overlook_feature(value, feature):
    if pd.isna(value) or value == "" or value.strip() == "Not Available":
        return "Not Available"
    elif feature in value:
        return True
    return False

# Apply the function to create new columns for each feature
df['Main Road'] = df['overlooking'].apply(lambda x: check_overlook_feature(x, "Main Road"))
df['Garden/Park'] = df['overlooking'].apply(lambda x: check_overlook_feature(x, "Garden/Park"))
df['Pool'] = df['overlooking'].apply(lambda x: check_overlook_feature(x, "Pool"))

```
